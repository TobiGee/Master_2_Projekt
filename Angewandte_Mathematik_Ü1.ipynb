## Aufgabe 1.1 
### Genauigkeit
Es sei p = 123 und q = 0,2. Bestimmen Sie die kleinere der beiden Lösungen von
$x^2 − 2px + q = 0$
mit den Maschinenzahlen zu b = 10 und m = 3 mithilfe der beiden folgenden ’Algorithmen’:
*  pq-Formel: $u_1=p-\sqrt{p^2-q}$
* pq-Formel + Erweitern: $u_1 = \frac{q}{p+\sqrt{p^2-q}}(=(p-\sqrt{p^2-q}) \cdot \frac{p+\sqrt{p^2-q}}{p+\sqrt{p^2-q}} )$ 

Vergleichen Sie die Ergebnisse mit den exakten Werten und interpretieren Sie die Resultate.


### Lösung:

Bestimmung der Gleitkomma Darstellungen von p und q:

P:

123 ist zur Basis 10 trivialer Weise 123.

Nun muss die Zahl Normalisiert werden:

$123=0,123 \cdot 10^3$

q:

Normalisieren führt zu:

$0,2 = 0,0002 \cdot 10^3$

#### Wir suchen nun die kleinere Lösung von:

$x^2-2 \cdot (0,123 \cdot 10^3) x + 0,0002 \cdot 10^3 = 0$


#### pq-Formel
$u_1=p-\sqrt{p^2-q}$

\begin{align}
&u_1 = 0,123 \cdot 10^3 - \sqrt{(0,123 \cdot 10^3)^2-0,0002 \cdot 10^3}&\text{Achtung: Rundung beim Quadrieren!}\\
&u_1 = 0,123 \cdot 10^3 - \sqrt{0,151 \cdot 10^5-0,0002 \cdot 10^3}& \text{Achtung: q hat keinen Einfluss!}\\
&u_1 = 0,123 \cdot 10^3 - \sqrt{0,151 \cdot 10^5}& \text{Achtung: q hat keinen Einfluss!}\\
&u_1 = 0,123 \cdot 10^3 - 0,123 \cdot 10^3& \text{Achtung: Rundung beim Radizieren}\\
&u_1 = 0&
\end{align}
Die Lösung $u_1 = 0$ entsteht dadurch, dass q bei der Subtraktion zu 0 gerundet wird und dadurch keinen Einfluss auf die Lösung hat.
#### pq-Formel + Erweitern
$u_1 = \frac{q}{p+\sqrt{p^2-q}}$

\begin{align}
&u_1 = \frac{0,0002 \cdot 10^3}{0,123 \cdot 10^3+\sqrt{(0,123 \cdot 10^3)^2-0,0002 \cdot 10^3}}&\text{Achtung: Rundung beim Quadrieren!}\\
&u_1 = \frac{0,0002 \cdot 10^3}{0,123 \cdot 10^3+\sqrt{0,151 \cdot 10^5-0,0002 \cdot 10^3}}&\text{Achtung: q hat keinen Einfluss!}\\
&u_1 = \frac{0,0002 \cdot 10^3}{0,123 \cdot 10^3+\sqrt{0,151 \cdot 10^5}}&\\
&u_1 = \frac{0,0002 \cdot 10^3}{0,123 \cdot 10^3+0,123 \cdot 10^3}\\
&u_1 = \frac{0,0002 \cdot 10^3}{0,246 \cdot 10^3}\\
&u_1 = 0,813 \cdot 10^{-3}
\end{align}
#### Exakte Lösung:
$u_1 =-8,1301081701865280218434506835555e^{-4}$

An dieser Aufgabe kann man sehr schön erkennen, weswegen beim Rechnen mit Gleitkommazahlen Fehler auftreten. Da bei Überschriten der Maximal verfügbaren Mantissenlänge gerundet werden muss, treten bei der Subtraktion im ersten Fall Fehler auf, die durch erweiter mit der nahhaften Null entfallen.

## Aufgabe 1.2 
### Kondition

Bestimmen Sie die Konditionierungszahlen der folgenden Rechenoperationen und entscheiden Sie
damit, in welchen Fällen die Operation gut-konditioniert ist.

a) Division: $f(x_1, x_2) = \frac{x_1}{x_2}$

b) Potenzieren: $f(x) = x^p$ für $x > 0$ und festes $p > 0$


### Lösung:
Definition Konditionszahl:

Ist f wie in Definition 1.8 und stetig differenzierbar, dann beschreibt die Konditionszahl (auch Verstärkungsfaktor)

$K_i =\frac{x_i}{f(x)}
\frac{∂f(x)}{∂x_i}$

die Kondition des entsprechenden Problems in Abhängigkeit von $x_i$. Im Idealfall ist $|Ki| ≤ 1$ für alle
$i$. Dann produziert der Algorithmus Ausgangsgrößen von der gleichen relativen Genauigkeit wie die
Eingangsgrößen.
Je kleiner die Konditionszahl, desto besser konditioniert ist das Problem.

#### Einsetzen der Funktionen

a) Division:

$K_1 =\frac{x_1}{f(x)}\frac{∂f(x)}{∂x_1}=x_2\frac{1}{x_2}=1$

$K_2 =\frac{x_2}{f(x)}\frac{∂f(x)}{∂x_2}=\frac{x_2^2}{x_1}(-\frac{x_1}{x_2^2}) = -1$

Da $|Ki| ≤ 1$ für alle $i$ gilt, ist das Problem gerade so gut konditioniert.

b) Potenzieren:

$K =\frac{x}{f(x)}\frac{∂f(x)}{∂x}=\frac{1}{x^{p-1}}px^{p-1} = p$

Beim Potenzieren entspricht die Konditionszahl dem Exponenten. Demzufolge ist das Problem nur für $p \leq 1$ gut Konditioniert.
Hier wäre ein Kurzer Gedanke zur Fehlerfortpflanzung bei Divisionen und Potenzierungen angebracht ;)



## Aufgabe 1.3 
### Entenjagd

In einem Entenschwarm fliegen zwischen 1000 und
10 000 Enten in einer regelmäßigen Dreiecksformation
(s. Skizze). Ein Jäger schießt nach den Enten, verfehlt sie aber.
Auf den Schreck organisieren sich die Enten neu, zu zwei exakt gleich
großen (kleineren) Schwärmen. Die Formation bildet jeweils wieder
ein regelmäßiges Dreieck.

Wieviele Enten fliegen im Schwarm?

### Numerische Lösung:
Gegeben:

Anzahl Enten $x$.

Bedingungen:
* Zahl $x$ ist Teil der Folge: $x_i = \sum_{n=1}^{i}n$.
* Zahl $\frac{x}{2}$ ist Teil der Folge: $x_i = \sum_{n=1}^{i}n$.

Vorgehen:
* Iteriere über Folge
* Speichere jede Iteration in Liste
* Halbiere Zahl und überprüfe, ob Zahl in Liste

Hinweise:
* Speicheroptimierung möglich durch Löschen aller Listeneinträge die kleiner als die Hälfte der aktiven Iteration sind.





```python
#Initialisieren der Liste
folge = [0]
#Loop bis 141, da das 142ste Folgenelement größer als 10000 ist.
for x in range(1, 141):
    folge.append(folge[-1]+x) 
    #Wenn hälfte der Zahl ebenfalls in Liste und Zahl größer als 1000
    if folge[-1]/2 in folge and folge[-1]>1000 :
        print("Es fliegen " + str(folge[-1]) + " Enten im Schwarm!")
```

## Aufgabe 1.4 
### Acht-Damen-Problem
Bestimmen Sie alle Lösungen des [Acht-Damen-Problems](https://de.wikipedia.org/wiki/Damenproblem) mit einem rekursiven Algorithmus.


```python
# Lösung aus Wikipedia:
# Erzeuge eine Liste von Lösungen auf einem Brett mit Reihen und Spalten.
# Eine Lösung wird durch eine Liste der Spaltenpositionen,
# indiziert durch die Reihennummer, angegeben.
# Die Indizes beginnen mit Null.
def damenproblem(reihen, spalten):
    if reihen <= 0:
        return [[]] # keine Dame zu setzen; leeres Brett ist Lösung
    else:
        return eine_dame_dazu(reihen - 1, spalten, damenproblem(reihen - 1, spalten))

# Probiere alle Spalten, in denen für eine gegebene Teillösung
# eine Dame in "neue_reihe" gestellt werden kann.
# Wenn kein Konflikt mit der Teillösung auftritt,
# ist eine neue Lösung des um eine Reihe erweiterten
# Bretts gefunden.
def eine_dame_dazu(neue_reihe, spalten, vorherige_loesungen):
    neue_loesungen = []
    for loesung in vorherige_loesungen:
        # Versuche, eine Dame in jeder Spalte von neue_reihe einzufügen.
        for neue_spalte in range(spalten):
            # print('Versuch: %s in Reihe %s' % (neue_spalte, neue_reihe))
            if kein_konflikt(neue_reihe, neue_spalte, loesung):
                # Kein Konflikte, also ist dieser Versuch eine Lösung.
                neue_loesungen.append(loesung + [neue_spalte])
    return neue_loesungen

# Kann eine Dame an die Position "neue_spalte"/"neue_reihe" gestellt werden,
# ohne dass sie eine der schon stehenden Damen schlagen kann?
def kein_konflikt(neue_reihe, neue_spalte, loesung):
    # Stelle sicher, dass die neue Dame mit keiner der existierenden
    # Damen auf einer Spalte oder Diagonalen steht.
    for reihe in range(neue_reihe):
        if (loesung[reihe]         == neue_spalte              or  # gleiche Spalte
            loesung[reihe] + reihe == neue_spalte + neue_reihe or  # gleiche Diagonale
            loesung[reihe] - reihe == neue_spalte - neue_reihe):   # gleiche Diagonale
                return False
    return True

for loesung in damenproblem(8, 8):
    print(loesung)
```
